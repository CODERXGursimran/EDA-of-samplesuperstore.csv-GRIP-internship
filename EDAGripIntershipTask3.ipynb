```python
#Importing all the libraries and data.
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
SData =pd.read_csv("SampleSuperstore.csv")
SData.head(10)
```

```python
# Understanding the dataset.
SData.info()
```

```python
# Counts the duplicate values on a column.
SData.duplicated().sum()
```

```python
# Dropping the duplicate values.
SData.drop_duplicates(inplace=True)
```

```python
SData.info()
```

```python
# By analysing the dataset, we can see that we dont require postal code column for furher analysis, therefore we will drop the column.
SData.drop(['Postal Code'], axis=1, inplace=True)
```

```python
# After removing the Additional column we can see now the data has 12 attributes and each containing 9977 samples.
SData.info()
```

```python
# Checking for Null Values.
SData.isna().sum()
```

#### No Null Values present in the Data set.

```python
# Now, finding correlation between attributes.
SData.corr()
```

#### We can see that the correlation exists only among the numerical attributes.

### Data Visualisation

```python
sns.heatmap(SData.corr(),annot=True)
```

#### Correlation is a statistical measure that expresses the extent to which two variables are linearly related (meaning they change together at a constant rate). Itâ€™s a common tool for describing simple relationships without making a statement about cause and effect.
The closer correlation value is to zero, the weaker the linear relationship.

Positive correlation values indicate a positive correlation, where the values of both variables tend to increase together.

Negative correlation values indicate a negative correlation, where the values of one variable tend to increase when the values of the other variable decrease.
##### For eg:- Here we can analyse that if sales increases, the profit increases by 48%. [ 1 represents strong positive correlation and -0.2 respresents max negative correlation for this dataset.]

```python
plt.figure(figsize=(15,15))
stpr=SData.groupby(['State'])['Profit'].sum().nlargest(50)
stpr.plot.bar()
```

##### The graph depicts that the store earns its major profit from California & New York whereas store is facing major losses in Ohio & Texas.

```python
SData.plot(kind="scatter",figsize=(12,6),x="Sales",y="Profit",c="Discount",s=20,fontsize=16,colormap='cividis')
plt.style.use('ggplot')
plt.show()
```

```python
sns.lineplot(x="Discount",y="Profit",label="Profit",data=SData)
plt.style.use('ggplot')
plt.legend()
plt.show()
```

#### We can clearly show that discount and profit are displaying a negative correlation,i.e. as Discount increases,the profit decreases.

```python
pls=SData.groupby('Region')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
pls[:].plot.bar(color=['orange','green'],figsize=(10,10))
plt.style.use('ggplot')
plt.title('Profit/Loss and Sales across the Region')
plt.xlabel('Region')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

#### As discounts are imposed on items, the sales are increasing but the  profits are decreasing tremendously.

```python
pls=SData.groupby('State')[['Sales','Profit']].sum().sort_values(by='Sales',ascending=False)
pls[:].plot.bar(color=['orange','green'],figsize=(10,10))
plt.title('Profit/Loss and Sales across the States')
plt.xlabel('States')
plt.ylabel('Profit/Loss and Sales')
plt.show()
```

## <b>Outcomes Derived After Exploratory Data Analysis (EDA) of the Retail store:</b>

<h4>1. States like 'California' and 'New York' have huge demand for the products sold by the store. Hence, the store should expand their operations in these regions.
<h4>2. The states like Dakota, Maine, West Virginia, District of Columbia show a steep decline in the sales attribute. Hence the store should study the market in these regions and increase sales of those products which are in high demand there to increase their sales and profits.
<h4>3. The 'Profit' generated by 'office supplies' category is more among other categories hence the store should 'increase the stock' of 'office supplies'. For 'furniture' and 'technology' related products, the company should only sell them in those regions where demand of 'furniture' and 'technology' related products is more.
<h4>4. The products like 'copiers' and and 'machines' contribute a very low percentage to sales of the store. But the store has mantained the stocks prehand, hence the store should only buy these stocks only when customer demand arises.
<h4>5. The standard of shipping prefered by around 60% of the store's customers is the standard class whereas other three shipping methods constitute to about 40%. Hence, if the store wants to increase the percentage of other shipping methods to enhance their profits, they have to change their pricing policies on these deliveries, enhance client experience and deliveries.
